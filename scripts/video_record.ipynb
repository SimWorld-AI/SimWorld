```python
from simworld.utils.video_recorder import VideoRecorder
from simworld.utils.vector import Vector
from simworld.communicator.communicator import Communicator
from simworld.communicator.unrealcv import UnrealCV
from simworld.agent.humanoid import Humanoid
```

```python
humanoid_1 = Humanoid(Vector(-600, 9000), Vector(1, 0)) # position here has default z-axis value 600
humanoid_2 = Humanoid(Vector(-650, 9100), Vector(1, 0)) # position here has default z-axis value 600
humanoid_3 = Humanoid(Vector(-700, 9200), Vector(1, 0)) # position here has default z-axis value 600

humanoids = [humanoid_1, humanoid_2, humanoid_3]
```

```python
communicator = Communicator(UnrealCV(port=9000, ip='127.0.0.1'))
```

```python
for humanoid in humanoids:
    communicator.spawn_agent(humanoid, '', position=[humanoid.position.x, humanoid.position.y, 600]) # reset humanoid position here to overwrite z-axis position
    communicator.humanoid_set_speed(humanoid.id, 100)
```

```python
def move_pattern(self, timestamp):
    """Generate a movement pattern for the humanoid at given timestamp."""
    actions = []
    step = timestamp

    # --- move forward ---
    self._apply_action( 'move_forward', None, timestamp, actions)

    return actions

video_recorder = VideoRecorder(
    communicator,
    humanoids,
    output_dir='./video_record',
    resolution=(1280, 720),
    fps=16.0,
    frame_num=50,
    move_pattern=None,
)
```

```python
video_recorder.record()
```
